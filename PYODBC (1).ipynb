```python
%sql SELECT count(winery) as number_of_wines, round(avg(price),2) as average_price, points from winemag_first150k group by points ORDER BY points DESC
```

```python
result2=%sql SELECT count(winery) as number_of_wines, round(avg(price),2) as average_price, points from winemag_first150k group by points ORDER BY points DESC
df2=result2.DataFrame()
```

```python
labels=df2['points']
N=len(df2)

ind=np.arange(N)

price=df2['average_price']

fig, ax1 = plt.subplots()
color='tab:red'

ax1.set_xlabel('Points',fontsize=15)
ax1.set_ylabel('Number of wines', color=color,fontsize=15)
ax1.bar(labels, df2['number_of_wines'], color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()
color = 'tab:blue'
ax2.set_ylabel('Avg. Price ($)', color=color,fontsize=15)  # we already handled the x-label with ax1
ax2.bar(labels, price, alpha=0.4, color=color)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()

plt.xticks(labels)
plt.gcf().set_size_inches(15, 10)
plt.show()
```
